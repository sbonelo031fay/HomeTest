# SR3 - EDA

```python
# importing necessary modules
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Reading CSV file
df = pd.read_csv(r"C:\Users\LENOVO\OneDrive\Desktop\bky6-3za5.csv")
```

### Scatter Plot

```python
age_of_is = df['age_of_is'].head(10)
structure_count = df['structure_count'].head(10)

plt.scatter(age_of_is, structure_count, color="green")
plt.title('Scatter of Age of Infrastructure vs Structure Count')
plt.xlabel("Age of Infrastructure")
plt.ylabel("Structure Count")
plt.grid(True)
plt.show()
```

### Bar Chart

```python
ward_councillor = df['ward_councillor'].head(4)
area_ha = df['area_ha'].head(4)

plt.figure(figsize = (8,5))
plt.bar(ward_councillor, area_ha, color='green', alpha = 0.7)
plt.title('Area HA by Ward Councillor')
plt.xlabel('Ward Councillor')
plt.ylabel('Area HA')
plt.show()
```

### Histogram

```python
age_of_is = df['age_of_is'].head(10)

plt.hist(age_of_is, bins = 4, color='green', alpha=0.7)
plt.title('Histogram of Age of Infrastructure in Settlements')
plt.xlabel('Age of Infrastructure')
plt.ylabel('Frequency')
plt.show()
```

## Advanced Plots
### Distribution Plot

```python
# Set Seaborn style
sns.set_style("whitegrid")

# Create the distribution plot with specific styling options
plt.figure(figsize=(8, 6))  # Set the figure size
sns.histplot(df['density_dwellings'].head(10), color='green', edgecolor='green')  # Customize plot color
plt.title('Distribution of Density Dwellings', fontsize=16)  # Set plot title and font size
plt.xlabel('Density Dwellings', fontsize=14)  # Set x-axis label and font size
plt.ylabel('Frequency', fontsize=14)  # Set y-axis label and font size
plt.xticks(fontsize=12)  # Sets font size for x-axis ticks
plt.yticks(fontsize=12)  # Sets font size for y-axis ticks
plt.grid(True)  # Show grid lines
##plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()  # Show the plot
```

### Swarm Plot

```python
plt.figure(figsize=(10, 6))  # Set the figure size
sns.swarmplot(x='ward_councillor', y='density_dwellings', data=df.tail(), size=8)  # Customize plot size and color palette
plt.title('Density Dwellings by Ward Councillor', fontsize=16)  # Set plot title and font size
plt.xlabel('Ward Councillor', fontsize=14)  # Set x-axis label and font size
plt.ylabel('Density Dwellings', fontsize=14)  # Set y-axis label and font size
plt.xticks(fontsize=12, rotation=45)  # Set font size and rotation for x-axis ticks
plt.yticks(fontsize=12)  # Set font size for y-axis ticks
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()  # Show the plot
```
